## Km driven 

```r
par(bg="white")
plot(log(df$km_driven), log(df$selling_price),
     xlab = "Log(Kilometers Driven)",
     ylab = "Log(Selling Price)",
     main = "Log-Log Relationship")
cor(log(df$km_driven), log(df$selling_price))
```

```r
cor(df$km_driven, df$selling_price, use = "complete.obs")
```

```r
model <- lm(selling_price ~ km_driven, data = df)

par(bg="white")
plot(df$km_driven, df$selling_price,
     xlab = "Kilometers Driven",
     ylab = "Selling Price",
     main = "Relationship between Kilometers Driven and Selling Price",
     pch = 19)
abline(model, col = "red", lwd = 2)
summary(model)
```

The results show a negative relationship between the number of kilometers driven and the selling price of a car. The correlation values are around −0.19 for the raw data and −0.29 after the log transformation, which means that when mileage increases, the selling price tends to decrease. Even if the relationship is not very strong, it is still clearly negative.

The scatter plot confirms this trend: most points are concentrated at low prices when the mileage is high, while cars with fewer kilometers generally have higher prices. The regression line goes downward, which shows the same pattern. According to the linear model, the coefficient of km_driven is −2.3850, meaning that the selling price decreases on average by about 2.39 units for each additional kilometer driven. This effect is statistically significant because the p-value is extremely small, so we are confident that mileage really influences the price.

However, the R-squared value is 0.0369, which means that kilometers driven explain only about 3.7% of the variation in selling price. In other words, mileage matters, but it is not the only factor. Other characteristics such as brand, year, condition, and model also play an important role in determining the value of a used car.

# Modelisation

```r
# model of vehicles 
df$model <- sapply(strsplit(as.character(df$name), " "), `[`, 2)
df$model <- factor(df$model)
```

## Train and test dataset

```r
vars <- c("selling_price",
          "year",
          "km_driven",
          "fuel",
          "seller_type",
          "transmission",
          "owner",
          "brand")

df_model <- df[, vars]

# character to factor 
for (v in names(df_model)) {
  if (is.character(df_model[[v]])) {
    df_model[[v]] <- as.factor(df_model[[v]])
  }
}

# Log of selling_price
df_model$log_price <- log(df_model$selling_price)
df_model$selling_price <- NULL
```

```r
set.seed(42)

index <- createDataPartition(df_model$log_price, p = 0.8, list = FALSE)

train_data <- df_model[index, ]
test_data  <- df_model[-index, ]
```

## Model fitting 

```r
# Cross Validation
ctrl <- trainControl(
  method = "repeatedcv",
  number = 5,      
  repeats = 3,     
  verboseIter = FALSE
)
```

### Linear regression 

```r
set.seed(42)
mod_lm <- train(
  log_price ~ .,
  data = train_data,
  method = "lm",
  trControl = ctrl,
  metric = "RMSE"
)
```

## Random Forest 

```r
set.seed(42)
mod_rf <- train(
  log_price ~ .,
  data = train_data,
  method = "ranger",
  trControl = ctrl,
  tuneLength = 5,      
  metric = "RMSE",
  importance = "impurity"
)
```

## XGBoost

```r
set.seed(42)
mod_xgb <- train(
  log_price ~ .,
  data = train_data,
  method = "xgbTree",
  trControl = ctrl,
  tuneLength = 5,
  metric = "RMSE",
  verbose = FALSE
)
```

## K-NN

```r
set.seed(42)
mod_knn <- train(
  log_price ~ .,
  data = train_data,
  method = "knn",
  trControl = ctrl,
  tuneLength = 10,   
  metric = "RMSE"
)
```

## LightGBM

```r
# One-hot encoding
train_mat <- model.matrix(log_price ~ . - 1, data = train_data)
test_mat  <- model.matrix(log_price ~ . - 1, data = test_data)

y_train <- train_data$log_price
y_test  <- test_data$log_price

dtrain <- lgb.Dataset(data = train_mat, label = y_train)

# Hyperparamaters grid
grid_lgb <- expand.grid(
  num_leaves      = c(15, 31),
  learning_rate   = c(0.05, 0.1),
  feature_fraction = 0.8,
  bagging_fraction = 0.8,
  bagging_freq     = 1
)

best_rmse  <- Inf
best_iter  <- NULL
best_param <- NULL

set.seed(42)
for (i in 1:nrow(grid_lgb)) {
  params <- list(
    objective        = "regression",
    metric           = "rmse",
    num_leaves       = grid_lgb$num_leaves[i],
    learning_rate    = grid_lgb$learning_rate[i],
    feature_fraction = grid_lgb$feature_fraction[i],
    bagging_fraction = grid_lgb$bagging_fraction[i],
    bagging_freq     = grid_lgb$bagging_freq[i]
  )
  
  cv <- lgb.cv(
    params = params,
    data = dtrain,
    nrounds = 500,
    nfold = 5,
    early_stopping_rounds = 20,
    verbose = -1
  )
  
  rmse_i <- min(unlist(cv$record_evals$valid$rmse$eval))
  iter_i <- which.min(unlist(cv$record_evals$valid$rmse$eval))
  
  if (rmse_i < best_rmse) {
    best_rmse  <- rmse_i
    best_iter  <- iter_i
    best_param <- params
  }
}

# Fit the model
mod_lgb <- lgb.train(
  params = best_param,
  data = dtrain,
  nrounds = best_iter,
  verbose = -1
)
```

## Performance on test data

```r
rmse <- function(obs, pred) sqrt(mean((obs - pred)^2))
mae  <- function(obs, pred) mean(abs(obs - pred))
r2   <- function(obs, pred) cor(obs, pred)^2
```

```r
pred_lm_log  <- predict(mod_lm,  newdata = test_data)
pred_rf_log  <- predict(mod_rf,  newdata = test_data)
pred_xgb_log <- predict(mod_xgb, newdata = test_data)
pred_knn_log <- predict(mod_knn, newdata = test_data)
pred_lgb_log <- predict(mod_lgb, test_mat)


# exp(log_price)
true_price   <- exp(y_test)

pred_lm  <- exp(pred_lm_log)
pred_rf  <- exp(pred_rf_log)
pred_xgb <- exp(pred_xgb_log)
pred_knn <- exp(pred_knn_log)
pred_lgb <- exp(pred_lgb_log)
```

```r
perf <- data.frame(
  Model = c("Linear Regression",
            "Random Forest",
            "XGBoost",
            "LightGBM",
            "k-NN"),

  RMSE = c(rmse(true_price, pred_lm),
           rmse(true_price, pred_rf),
           rmse(true_price, pred_xgb),
           rmse(true_price, pred_lgb),
           rmse(true_price, pred_knn)),

  MAE = c(mae(true_price, pred_lm),
          mae(true_price, pred_rf),
          mae(true_price, pred_xgb),
          mae(true_price, pred_lgb),
          mae(true_price, pred_knn)),

  R2 = c(r2(true_price, pred_lm),
         r2(true_price, pred_rf),
         r2(true_price, pred_xgb),
         r2(true_price, pred_lgb),
         r2(true_price, pred_knn))
)

perf[order(perf$RMSE), ]
```

```r
par(bg = "white", mfrow = c(1, 3), mar = c(8, 4, 4, 1))

# --- RMSE ---
barplot(
  perf$RMSE,
  names.arg = perf$Model,
  las = 2,
  col = "lightblue",
  main = "RMSE by Model",
  ylab = "RMSE",
  cex.names = 0.8
)

# --- MAE ---
barplot(
  perf$MAE,
  names.arg = perf$Model,
  las = 2,
  col = "orange",
  main = "MAE by Model",
  ylab = "MAE",
  cex.names = 0.8
)

# --- R^2 ---
barplot(
  perf$R2,
  names.arg = perf$Model,
  las = 2,
  col = "lightgreen",
  main = expression(R^2 ~ "by Model"),
  ylab = expression(R^2),
  ylim = c(0, 1),
  cex.names = 0.8
)

par(mfrow = c(1,1))
```

Several machine learning models were trained to predict the selling price of used cars. Their performance was evaluated on the test set using three metrics: RMSE (Root Mean Squared Error), MAE (Mean Absolute Error), and R² (coefficient of determination). The results are shown in the table.

The Linear Regression model achieved the best performance, with the lowest RMSE (259,369.4) and the highest R² value (0.8416). This means that the linear model explains around 84% of the variability in car prices and provides the most accurate predictions among all tested algorithms. Surprisingly, this simple model outperformed more complex models such as XGBoost and LightGBM, even though these advanced boosting techniques are usually known for strong predictive power. This suggests that the relationship between the predictors and the log of selling price is mostly linear in this dataset.

XGBoost and LightGBM followed closely behind, with similar RMSE and $R^2$ values, meaning they are still strong options for prediction tasks. The Random Forest model performed moderately well but did not reach the same accuracy level, while the k-NN model obtained the worst results, with a very high RMSE and a much lower R², making it unsuitable for this problem.

For future work, feature engineering or interaction effects could be explored to improve the performance of advanced models such as XGBoost or LightGBM. We could create new variables such as age of vehicule, price depreciation rate (if we get the new vehicles prices) or look into the interactions between brand and transmission type for instance. It may help capture non linear relationships that the current models can't fully detect.
